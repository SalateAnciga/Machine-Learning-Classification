```python
import pandas as pd 
import numpy as np
import matplotlib as pyplot
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal','pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset[['classification_yes']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],'penalty':['none', 'elasticnet', 'l1', 'l2']}
            
```

```python
grid=GridSearchCV(LogisticRegression(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
```

```python
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print(re)
```

```python
Grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,Grid_prediction)
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
clf= classification_report(y_test,Grid_prediction)
print(clf)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,Grid_prediction,average='weighted')
print("The best parameter{}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc=roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
print(roc)
```
